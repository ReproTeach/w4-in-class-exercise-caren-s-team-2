## 2. Preprocess the data (from week3 exercise)

1. Rename all columns
2. Drop nan
3. Drop duplicates

```python
# write your code here, you can add more cells if needed

import pandas as pd

# read your data
df = pd.read_csv('pre-course_survey.csv')
    
# rename the columns
df.rename(columns={'column1': 'new_name', 'column2': 'new_name2'}, inplace=True)
old_names = df_csv.columns.values
new_names = ['timestamp', 'python', 'datascience', 'git', 'jupyter', 'reproducibility', 'nlp', 'socialmedia', 'goals', 'skills', 'concerns']
replace_columns = dict(zip(old_names,new_names))
df_csv.rename(columns=replace_columns, inplace=True)
df_csv
```

```python
# drop nan
df_csv.dropna()
    
# drop duplicates
df_csv.drop_duplicates()
    
# recode yes/no to 1/0
df_csv = df_csv.replace({'Yes':1,'No':0})
df_csv
    
# return df_csv
```
